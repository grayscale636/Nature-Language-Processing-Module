```python
import pandas as pd
#input data dari local file
data = pd.read_csv('hasil_saya.csv')
data.head()
```

```python
data.info()
```

```python
indeks = 0
data = data.drop(data.columns[indeks], axis=1)
```

```python
data.head()
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import re

processed_comments = []

for sentence in data['content']:
    
    # Mengubah ke lowercase
    processed_comment = sentence.lower()
    
    # Substituting multiple spaces with single space
    processed_comment = re.sub(r'\s+', ' ', processed_comment, flags=re.I)
    
    # Menghilangkan karakter spesial
    processed_comment = re.sub(r'\W', ' ', processed_comment)

    # Menghilangkan stopword
    factory = StopWordRemoverFactory()
    stopwords = factory.get_stop_words()
    temp = [t for t in re.findall(r'\b[a-z]+-?[a-z]+\b', processed_comment) if t not in stopwords]
    processed_comment = ' '.join(temp)
    
    processed_comments.append(processed_comment)
    
# Output data Preprocessing
processed_comments
```

```python
data['content'] = pd.DataFrame(processed_comments, columns=['review'])
```
